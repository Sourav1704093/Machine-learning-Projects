```python
import pandas as pd
import numpy as np
import itertools
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
df = pd.read_csv('news.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
#get the labels from the dataframe
labels = df.label
labels.head()
```

```python
#spliting dataset insto train and test set
X_train,X_test,y_train,y_test = train_test_split(df['text'],labels,test_size = 0.3,random_state = 8)
```

```python
#initializing a TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words = 'english',max_df = 0.6)
```

```python
print(tfidf_vectorizer.get_feature_names())
```

```python
#fit and transform train set and transform test set
tfidf_train = tfidf_vectorizer.fit_transform(X_train)
tfidf_test = tfidf_vectorizer.transform(X_test)
```

```python
#Initialize passive aggressive classifier
pagc= PassiveAggressiveClassifier(max_iter = 50)
pagc.fit(tfidf_train,y_train)
```

```python
#predict on the test set and calculate accuracy
y_pred = pagc.predict(tfidf_test)
score = accuracy_score(y_test,y_pred)
```

```python
print(score)
```

```python
print(f'Accuracy: {round(score*100,2)}%')
```

```python
#build the confusion matrix
confusion_matrix(y_test,y_pred,labels = ['FAKE','REAL'])
```
